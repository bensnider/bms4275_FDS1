1. Write a function to convert the data file "018_HW_2x2_data.csv" from wide to "tidy" long format. Your function can assume that the input data consist of 4 columns (like our rat data), but should handle any number of rows (i.e. number of observations per cell).

```python
def tidy_data(filename) :
    import pandas as pd
    import numpy as np
    data_input = pd.read_csv(filename)
    data_numpy = data_input.to_numpy()
    rows, columns = data_numpy.shape
    total_obsv = rows*columns
    data_reshaped = np.reshape(data_numpy, (total_obsv, 1), order = 'F')
    data_reshaped = np.squeeze(data_reshaped)
    cond_a = pd.Series(['condition_a_level_1', 'condition_a_level_2'])
    cond_a = cond_a.repeat(rows)
    cond_a = pd.concat([cond_a]*2, ignore_index=True) 
    cond_b = pd.Series(['condition_b_level_1', 'condition_b_level_2'])
    cond_b = cond_b.repeat(2*rows)
    cond_b = cond_b.reset_index(drop = True)
    tidy_df =pd.DataFrame({"data": data_reshaped, "condition a": cond_a, "condition b": cond_b})
    return tidy_df
```

```python
tidy_data('datasets/018_HW_2x2_data.csv')
```
