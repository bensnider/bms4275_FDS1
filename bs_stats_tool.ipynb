### README

To run the tool just execute the cell of code below and follow all user inputs. The tool should (hopefully) work smoothly!

```python
### Creating the actual tool
import bs_stats_module as stat
enter_tool = True
print("Hello! Welcome to Ben's stat tool!\n ")
valid_select = [1, 2, 3, 4, 5, 6, 7, 8]
while enter_tool == True :
    try :   
        print("Please select an option below: \n[1] Input data set \n[2] Calculate mean \n[3] Calculate standard deviation \n[4] Calculate variance \n[5] Calculate standard error \n[6] Calculate z-score for a new observation \n[7] Calculate summary of the statistics \n[8] Quit stat tool \n")
        menu_select = float(input("Please select the number of the option you would like: \n"))
        if menu_select == 1:
            done_test = True
            data = []
            while done_test == True:
                data_input = input("Enter each observation in your dataset, when finished type 'done': \n")
                if data_input.lower() == 'done' :
                    done_test = False
                    break
                num = float(data_input)
                data.append(num)
        if menu_select == 2:
            print(f"\nMean: {stat.mean(data)} \n")
        if menu_select == 3:
            print(f"\nStandard Deviation: {stat.sd(data)} \n")
        if menu_select == 4:
            print(f"\nVariance: {stat.var(data)} \n")
        if menu_select == 5:
            print(f"\nStandard Error: {stat.se(data)} \n")
        if menu_select == 6:
            new_value = float(input("Input new observation: \n"))
            print(f"\nZ-score of {new_value} : \n{stat.z_score(data, new_value)} \n")
        if menu_select == 7:
            print(f"\n{stat.summ(data)} \n")
        if menu_select == 8 :
            print("\nThanks for using Ben's stat tool!")
            enter_tool = False
        menu_test = menu_select in valid_select
        if menu_test == False :
            print("\nUh oh! This is not a valid menu selection, please try again! \n")
    except :
        print("\nOh no! Looks like something has gone wrong. Please try again.\n")
```
